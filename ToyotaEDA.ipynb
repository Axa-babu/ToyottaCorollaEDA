# Boxplot of Price, Km column

A boxplot is a graph that gives a visual indication of how a data set's 25th percentile, 50th percentile, 75th percentile, minimum, maximum and outlier values are spread out and compare to each other.

```python
sns.boxplot(x=df['Price'],palette='Paired')
sns.despine()
plt.show()
```

```python
sns.boxplot(x=df['Km'],palette='Set1')
sns.despine()
plt.show()
```

```python
q1=df["Price"].quantile(0.25)
q3=df["Price"].quantile(0.75)
iqr=q3-q1
lower_limit = q1 - 6.5*iqr

upper_limit = q3 + 0.5*iqr

lower_limitoutliers = df[df['Price'] < lower_limit]

upper_limitoutliers = df[df['Price'] > upper_limit]
new_df= df[(df['Price'] < upper_limit)&(df['Price'] >lower_limit)]

import matplotlib.pyplot as plt

import seaborn as sns

sns.set(rc={"figure.figsize": (6, 6)})

sns.boxplot(x=new_df["Price"])

plt.legend(loc="best")
```

```python
q1=df["Km"].quantile(0.25)
q3=df["Km"].quantile(0.75)
iqr=q3-q1
lower_limit = q1 - 6.5*iqr

upper_limit = q3 + 0.5*iqr

lower_limitoutliers = df[df['Km'] < lower_limit]

upper_limitoutliers = df[df['Km'] > upper_limit]
new_df= df[(df['Km'] < upper_limit)&(df['Km'] >lower_limit)]

import matplotlib.pyplot as plt

import seaborn as sns

sns.set(rc={"figure.figsize": (6, 6)})

sns.boxplot(x=new_df["Km"])

plt.legend(loc="best")
```
